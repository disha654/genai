```python
import gensim.downloader as api
from scipy.spatial.distance import cosine

# Load pre-trained word2vec model
model = api.load("word2vec-google-news-300")

def word_vector_arithmetic(word1, word2, operation):
    # Check if both words are in the model vocabulary
    if word1 not in model.key_to_index or word2 not in model.key_to_index:
        return "One or both words not in vocabulary."
    
    vec1, vec2 = model[word1], model[word2]

    # Perform the vector arithmetic
    if operation == '+':
        result_vec = vec1 + vec2
    elif operation == '-':
        result_vec = vec1 - vec2
    else:
        return "Invalid operation"

    # Get the most similar word
    similar_word = model.similar_by_vector(result_vec, topn=1)[0][0]
    return similar_word

# Example usage
word1 = "queen"
word2 = "person"
operation = "+"
result = word_vector_arithmetic(word1, word2, operation)
print(f"Result of '{word1} {operation} {word2}': {result}")
```
