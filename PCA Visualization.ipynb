```python
import gensim.downloader as api
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Load GloVe model
model = api.load('glove-wiki-gigaword-100')

# Use lowercase versions of words
words = ['computer', 'internet', 'software', 'hardware', 'data', 'ai', 'robot', 'network', 'cloud', 'security']

# Filter to include only words present in model
valid_words = [word for word in words if word in model]
vectors = [model[word] for word in valid_words]

# Reduce dimensions to 2D
pca = PCA(n_components=2)
reduced = pca.fit_transform(vectors)

# Plot
plt.figure(figsize=(8, 6))
for i, word in enumerate(valid_words):
    plt.scatter(*reduced[i])
    plt.text(reduced[i][0]+0.01, reduced[i][1]+0.01, word)
plt.title('PCA Visualization of Tech Word Embeddings')
plt.grid(True)
plt.show()
```

```python
# Input word
word = "robot"
similar_words = model.most_similar(word, topn=5)

# Output
for sim_word, score in similar_words:
    print(f"{sim_word} ({score:.2f})")
```
